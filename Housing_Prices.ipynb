### General Information About Data:
#### In This Part, We Will Need To:
* ##### Read The Data.
* ##### Have A Look At The Data's General Information.
* ##### Have A Look At The Data's Ranges.
* ##### Have A Look At The Data's Missing Values, Or Duplicated Instances.
* ##### Have A Look At The Data's Statistical Description.

```python
houses_data = pd.read_csv("Data/train.csv")
houses_data
```

```python
houses_data.drop(['Id'], axis= 1, inplace= True)
```

```python
print("Data Size:",houses_data.shape)
print("-"* 30)
print("Data Columns:",list(houses_data.columns))
print("-"* 30)
print("Number of Examples 'N'=",houses_data.shape[0])
print("Number of Columns=",houses_data.shape[1])
print("-"* 30)
print("Data Check For Any Duplicates:",houses_data.duplicated().any())
print("-"* 30)
print("About Dataset:")
houses_data.info()
```

```python
numeric_data = houses_data.select_dtypes(include=np.number)
numeric_data
```

```python
cateogrical_data = houses_data.select_dtypes(exclude=np.number)
cateogrical_data
```

```python
Columns = list(numeric_data.columns)
discrete_cols = list()
continous_cols = list()
for column in Columns:
    n_unique = len(np.unique(numeric_data[column]))
    # This if statment by ChatGPT.
    # if the number of unique values is less than or equal to 5% of the total number of values,
    # consider the column as discrete
    if n_unique <= 0.05 * len(numeric_data[column]):
        discrete_cols.append(column)
    else:
        continous_cols.append(column)
print('Columns considered as discrete variables:')
print(discrete_cols)
print('-'*30)
print('Columns considered as continous variables:')
print(continous_cols)
discrete_data = numeric_data[discrete_cols]
continous_data = numeric_data[continous_cols]
```

```python
for column in Columns:
    print("'" + column + "'", "Range Of Values:",numeric_data[column].min(),"-",numeric_data[column].max())
    print("-"* 30)
```

```python
duplicated_rows = houses_data[houses_data.duplicated()]
print("Number of duplicate rows: ", duplicated_rows.shape[0])
```

```python
print('Number of Total missing/null values:')
missing_check = houses_data.isnull().sum()

print(missing_check.iloc[np.nonzero(missing_check.to_numpy())])
```

```python
houses_data.describe().round(5)
```

### Visualizations Against The Target And Explorations:
#### In This Part, We Will Do:
* ##### Read The Data.
* ##### Have A Look At The Data's Correlation Heatmap.
* ##### Have A Look At The Data's Price Vs Features In The Data From Categorical And Numerical Data.

```python
sns.heatmap(round(numeric_data.corr(),2), cmap="BrBG", annot= True)
plt.rcParams["figure.figsize"] = (30, 30)
plt.show()
```

```python
default_figsize = mpl.rcParams['figure.figsize']
plt.rcParams['figure.figsize'] = default_figsize
```

```python
num_col = continous_data.shape[1]
indices = np.arange(0, num_col, 3)
for i in indices:
    if i!= indices[-1]:
        fig, axes = plt.subplots(2, 3, gridspec_kw={"height_ratios": (0.7, 0.3)}, figsize=(15, 6))
        for (feat, plot_col) in zip(list(continous_data.columns)[i:i+3], range(3)):  # pairing feature with column in plot
            axes[0, plot_col].hist(houses_data[feat], bins=50)
            axes[0, plot_col].set_title("Histogram and boxplot for {}".format(feat))
            sns.despine(ax=axes[0, plot_col])
            sns.boxplot(data=houses_data[feat], orient="h", ax=axes[1, plot_col], fliersize=2)
            axes[1, plot_col].set(yticks=[])
            sns.despine(ax=axes[1, plot_col], left=True)

        for ax in axes.flatten():
            ax.set_xlabel("")
        plt.show()
```

```python
for i in indices:
    if i != indices[-1]:
        sns.pairplot(data=continous_data, y_vars="SalePrice",x_vars=continous_data.columns[i:i+3],plot_kws={'color':'black','edgecolor':'green','alpha':0.5})
        plt.show()
```

```python
for i in range(len(discrete_data.columns)):
    sns.barplot(data=discrete_data, y=continous_data["SalePrice"], x=discrete_data.columns[i])
    plt.show()
```

```python
for i in range(len(discrete_data.columns)):
    sns.barplot(data=cateogrical_data, y=continous_data["SalePrice"], x=cateogrical_data.columns[i])
    plt.show()
```
